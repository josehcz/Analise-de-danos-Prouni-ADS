### Mensalidade das faculdades com nota minima (450)

```python
df.query('nota_integral_ampla == "450.0"')[['cidade_busca','universidade_nome','mensalidade']].sort_values(by='mensalidade').head(100)
```

### Faculdades com nota de corte maior que 650

```python
df.query('nota_integral_ampla > 650.0')[['cidade_busca','universidade_nome','mensalidade','nota_integral_ampla']].sort_values(by='mensalidade').head(100)
```
